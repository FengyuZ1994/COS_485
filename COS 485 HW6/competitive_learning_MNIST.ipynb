```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg
from sklearn.datasets import fetch_mldata
```

```python
mnist = fetch_mldata('MNIST original', data_home='data')
train = np.array(mnist.data / 255.)
trainlabels = mnist.target
```

```python
train.shape
```

```python
data_size = train.shape[0]
input_size = train.shape[1]
```

```python
def train_all():
    
    index = np.random.permutation(np.arange(data_size))
    X = train[index]
    Y = trainlabels[index]
    
    ETA = 0.04
    n = input_size
    k = 10
    
    #initialize the weights of 10 classes
    W = np.mean(X, axis = 0) + 0.01*np.random.randn(k, n)
    
    #loop of the whole dataset
    for i in range(data_size):
        diff = linalg.norm(X[i] - W, axis = 1) 
        W[diff.argmin()] += ETA*(X[i] - W[diff.argmin()])     
    
    #Show the figures
    fig = plt.figure()
    for i in range (10):
        fig.add_subplot(2, 5, i+1)
        plt.imshow(W[i].reshape(28, 28))
    plt.show()
    

def train_two():
    
    index_two = (trainlabels == 2.0)
    X = train[index_two]
    Y = trainlabels[index_two]
    
    ETA = 0.04
    n = input_size
    k = 5
    
    #initialize the weights of 10 classes
    W = np.mean(X, axis = 0) + 0.01*np.random.randn(k, n)
    
    #loop of the whole dataset
    for i in range(len(Y)):
        diff = linalg.norm(X[i] - W, axis = 1) 
        W[diff.argmin()] += ETA*(X[i] - W[diff.argmin()])     
    
    #Show the figures
    fig = plt.figure()
    for i in range (5):
        fig.add_subplot(2, 5, i+1)
        plt.imshow(W[i].reshape(28, 28))
    plt.show()
```

```python
train_all()
```

Competitive Learning Lagorithm on all images with k = 10 : 

They look like the 10 digits, except that 4 and 5 are missing. The results depends on the initialization which would affect the input images assigned to each weight vector. 

Of course they look like the samples in the dataset due to the fact that the output weight vectors are basically the mean vectors of assigned input images.


```python
train_two()
```

The algorithm has learned several mean values of 'two' images. These are the most popular mean vectors. In other words, the algorithm learned five different styles of 'two' images.
